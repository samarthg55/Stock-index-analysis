# Main Class and Sub-Classes

```python
class Index:
    def __init__(self,date,index_name,company,type,rating,price,fluct_rate):
        self.date = date
        self.index_name = index_name
        self.company = company
        self.type = type
        self.rating = rating
        self.price = price
        self.fluct_rate = fluct_rate
        self.unique_id = f"{date}_{index_name}"
    def __str__(self):
        return f"{self.unique_id},{self.date},{self.index_name},{self.company},{self.type},{self.rating},{self.price},{self.fluct_rate}"

class LargeCapIndex(Index):
    def __init__(self,date, index_name, company, rating, price, fluct_rate):
        super().__init__(date, index_name, company, "LargeCap", rating, price, fluct_rate)

class NonLargeCapIndex(Index):
    def __init__(self,date, index_name, company, rating, price, fluct_rate):
        super().__init__(date, index_name, company, "non-LargeCap", rating, price, fluct_rate)

class SNP(LargeCapIndex):
    def __init__(self,date, index_name, rating, price, fluct_rate):
        super().__init__(date, index_name, "SNP", rating, price, fluct_rate)

class DJ(NonLargeCapIndex):
    def __init__(self,date, index_name, rating, price, fluct_rate):
        super().__init__(date, index_name, "DJ", rating, price, fluct_rate)

class NASDAQ(NonLargeCapIndex):
    def __init__(self,date, index_name, rating, price, fluct_rate):
        super().__init__(date, index_name, "NASDAQ", rating, price, fluct_rate)
```

We will Replace NaN or Null of Date with mode(the most frequently occurring date in that column)

index_data['Date'].mode()[0]: This part calculates the mode (most frequently occurring value) of the 'Date' column. The .mode() method returns a Pandas Series containing the mode(s) of the 'Date' column. [0] is then used to access the first mode in case there are multiple modes.

```python

index_data['Date'] = index_data['Date'].fillna(index_data['Date'].mode()[0])
```

We will Replace NaN or Null in Rating with Aggregate mean at level of granularity of 'Rating'
by filling missing values in the 'Rating' column based on the mean (average) value of each group defined by the 'Index_Type' column in the Pandas DataFrame index_data

```python

index_data['Rating'] = index_data.groupby('Index_Type')['Rating'].transform(lambda x: x.fillna(x.mean()))
```

We will Replace NaN or Null in Price with Aggregate median at level of granularity of 'index_name'

.transform(lambda x: x.fillna(x.median())): This is a transformation operation applied to each group independently. The lambda function lambda x: x.fillna(x.median()) is used to fill missing values within each group with the median value of that group.

x represents each group,
x.fillna(x.median()) fills missing values in the 'Price' column of each group with the median of that group.

```python

index_data['Price'] = index_data.groupby('Index_Name')['Price'].transform(lambda x: x.fillna(x.median()))
```


We will Replace NaN or Null in Fluct_Rate with Aggregate mean at level of granularity of 'index_name' by filling missing values in the 'Fluct_Rate' column based on the mean value of each group defined by the 'Index_Name' column in the Pandas DataFrame index_data

.transform(lambda x: x.fillna(x.mean())): This is a transformation operation applied to each group independently. The lambda function lambda x: x.fillna(x.mean()) is used to fill missing values within each group with the mean value of that group.

x represents each group.
x.fillna(x.mean()) fills missing values in the 'Fluct_Rate' column of each group with the mean of that group.


```python
index_data['Fluct_Rate'] = index_data.groupby('Index_Name')['Fluct_Rate'].transform(lambda x: x.fillna(x.mean()))
```

```python
import pandas as pd

# Load data from the CSV file
file_path = 'data.csv'
df = pd.read_csv(file_path)

# Calculate mean values
mean_values = df.groupby('company')['price'].mean().reset_index()

# Print mean values
print(mean_values)
```

```python

# Specify the desired order of companies
company_order = ['NASDAQ', 'SNP', 'DJ']

# Calculate mean prices for each company
mean_prices = df.groupby('company')['price'].mean().reset_index()

# Sort the DataFrame based on the desired order
mean_prices['company'] = pd.Categorical(mean_prices['company'], categories=company_order, ordered=True)
mean_prices = mean_prices.sort_values('company')

# Increase the figure size for better readability
plt.figure(figsize=(10, 6))

# Plot the bar chart with the specified order
plt.bar(mean_prices['company'], mean_prices['price'], color=['red', 'orange', 'yellow'])
plt.title('Average Price per Company')
plt.xlabel('Company')
plt.ylabel('Average Price')

# Adjust y-axis range and ticks for better readability
plt.ylim(198, max(mean_prices['price']) + 1)
plt.yticks(range(198, int(max(mean_prices['price'])) + 1, 1))

# Show the plot
plt.show()

```

Market Dominance vs. Transaction Value:

NASDAQ's higher number of entries sold indicates a larger market share or higher trading volume compared to SNP and DJ. This suggests that NASDAQ is more actively traded in terms of the number of transactions.

On the other hand, DJ, despite having the fewest entries sold, has the highest average transaction value. This could mean that DJ, although not as frequently traded, commands higher per-entry transaction values, potentially indicating a market with higher-value assets.

Central Tendency: The centerline of the box (median) is around 20.75. This represents the middle value of the dataset. It's not affected by extreme values and provides a measure of central tendency.

Outliers: The boxplot can also give insights into potential outliers. Any data points beyond the whiskers (lines extending from the box) might be considered outliers. In this case, you may want to investigate if there are any extreme values in the 'Close' prices.

Distribution Shape: The shape of the boxplot provides information about the distribution of the data. A symmetric box suggests a relatively symmetric distribution, while skewness or asymmetry might be indicated if one whisker is longer than the other.
